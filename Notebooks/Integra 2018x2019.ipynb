**COMO O PANDAS CALCULA A PORCENTAGEM**

faz a regra de 3 e obtem o valor

1.579963 - 100%

0.15242494226327943 - x

1.579963x = 15.242494226327944

x = 15.242494226327944/1.579963

9.64737416403292
