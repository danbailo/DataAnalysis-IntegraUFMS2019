### **RELAÇÃO DE TRABALHO SUBMETIDO/QTD DE ALUNOS**

```python
relacao = df['Unidade'].value_counts()/qtd['unidade'].value_counts()
```

```python
relacao.dropna().sort_values(ascending=False).plot.pie(subplots=True, legend=True, figsize=(10, 10), shadow=True, autopct='%1.2f%%', colors=colors)
```
